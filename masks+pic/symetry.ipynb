```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology

def calculate_symmetry_score(mask_image):
    # Erode the mask to find the border
    struct_el = morphology.disk(1)
    mask_eroded = morphology.binary_erosion(mask_image, struct_el)
    mask_border = mask_image - mask_eroded
    
    # Compute perimeter and area
    computed_perimeter = np.sum(mask_border)
    computed_area = np.sum(mask_image)
    
    # Calculate symmetry score
    symmetry_score = min(computed_perimeter, computed_area) / max(computed_perimeter, computed_area)
    
    # Convert symmetry score to a scale of 0-3
    scale = int(symmetry_score * 3)
    
    return scale

# Example usage
if __name__ == "__main__":
    # Load the mask image
    mask_image = cv2.imread('PAT_492_1784_50_mask.png', cv2.IMREAD_GRAYSCALE)
    
    # Check if mask_image is loaded correctly
    if mask_image is None:
        print("Error: Failed to load the mask image.")
```
